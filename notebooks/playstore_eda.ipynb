**1️⃣ What are the top 10 app categories on the Google Play Store?**

```python
df['Category'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
top_category=df["Category"].value_counts().head(10)
font1={"family":"serif",'size':15,'color':'darkred'}
font2={"family":'sans-serif','size':12,'color':'red'}
plt.ylabel("Category",fontdict=font2)
plt.xlabel("Number of Apps",fontdict=font2)
plt.title('Top 10 App Categories',fontdict=font1)
plt.barh(top_category.index,top_category.values,color='orange')
plt.savefig("../images/top_10_categories.png",bbox_inches='tight')
plt.show()
```

**📌 Insight:
The Family and Game categories dominate the Play Store, suggesting high competition among apps in these categories**.


**2️⃣ What is the distribution of app ratings?**

```python
from matplotlib import pyplot as plt
font1={"family":"serif",'size':15,'color':'darkred'}
font2={"family":'sans-serif','size':12,'color':'red'}
plt.ylabel("Number of Apps",fontdict=font2)
plt.xlabel("Rating",fontdict=font2)
plt.title('Distribution of App Ratings',fontdict=font1)
plt.xticks([1,2,3,4,5])
plt.hist(df["Rating"].dropna(),histtype='stepfilled',edgecolor='g',facecolor='orange')
plt.savefig('../images/distribution_app_ratings.png',bbox_inches='tight')
plt.show()
```

**📌 Insight:
Most apps are rated between 4.0 and 4.5, indicating generally positive user feedback.**

**3️⃣ Which app categories have the highest number of installs?**

```python
import matplotlib.pyplot as plt
top_installs=df.groupby("Category")["installs_1"].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
plt.bar(top_installs.index,top_installs.values,color="orange")
font1={"family":"serif",'size':15,'color':'darkred'}
font2={"family":'sans-serif','size':12,'color':'red'}
plt.ylabel("Number of Installs (In Billions)",fontdict=font2)
plt.xlabel("Category",fontdict=font2)
plt.title('Total Installs per App Category',fontdict=font1)
plt.xticks(rotation=45, ha='right')
plt.savefig('../images/highest_installs.png',bbox_inches='tight')
plt.show()
```

**📌 Insight:
Categories like Game and Communication have the highest total installs, showing strong user demand and engagement.**

**4️⃣ What is the relationship between app ratings and number of reviews?**

```python
import matplotlib.pyplot as plt

df_scatter = df[["Rating", "Reviews"]].dropna()
df_scatter = df_scatter[df_scatter["Rating"].between(1, 5)]
df_scatter["Reviews_million"] = df_scatter["Reviews"] / 1e6

font1 = {"family": "serif", "size": 15, "color": "darkred"}
font2 = {"family": "sans-serif", "size": 12, "color": "black"}

plt.figure(figsize=(14,6))

plt.scatter(
    df_scatter["Rating"],
    df_scatter["Reviews_million"],
    c=df_scatter["Rating"],
    cmap="rainbow",
    alpha=0.6,
    s=20
)

plt.xlabel("Rating", fontdict=font2)
plt.ylabel("Number of Reviews (Millions)", fontdict=font2)
plt.title("Relationship Between App Ratings and Reviews", fontdict=font1)

plt.ticklabel_format(style='plain', axis='y')

cbar = plt.colorbar()
cbar.set_label("Rating")

plt.savefig("../images/relation_between_ratings_reviews.png", bbox_inches="tight")
plt.show()
```

**📌 Insight:
Apps with higher ratings tend to receive more reviews, showing a positive relationship between user satisfaction and engagement.**
